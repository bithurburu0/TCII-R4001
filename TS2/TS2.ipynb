#### Simulación simbólica
En el siguiente código Python, a través del uso de la biblioteca $SymPy$ se verifica por método de simulación simbólica la función transferencia obtenida en el punto **1)**.

```python
# variables simbólicas
s = sp.symbols('s', complex=True)
Vi, Vo, Vxa, Vxb = sp.symbols("Vi, Vo, Vxa, Vxb")
G1, G2, G3, G4, C = sp.symbols("G1, G2, G3, G4, C")

# Sistemas de ecuaciones del modelo ideal, negativamente realimentado
aa = sp.solve([
                G1*Vi + G3*Vo + G2*Vxa + s*C*Vxa, 
                G3*Vxa + s*C*Vxb,
                G4*Vxb + G4*Vo
                ], 
                [Vi, Vo, Vxa, Vxb])
T1 = aa[Vo]/aa[Vi]

num, den = sp.fraction(sp.simplify(sp.expand(T1)))
num = sp.Poly(num,s)
den = sp.Poly(den,s)
num1, den1, w_on, Q_n, w_od, Q_d, k = parametrize_sos(num, den)

print_subtitle('Transferencia obtenida')

display(Math( r' \frac{V_o}{V_i} = ' + sp.latex(sp.Mul(k,num1/den1, evaluate=False)) ))

print_subtitle('Parámetros')

display(Math( r' k = ' + sp.latex(k) ))

display(Math( r' \omega_0 = ' + sp.latex(w_od) ))

display(Math( r' Q = ' + sp.latex(Q_d) ))
```

#### Simulación Python Butterworth de orden 2
En el siguiente código Python se realiza la simulación de la función transferencia normalizada obtenida en el punto **3)**, para un Filtro Pasa-Bajos Butterworth de orden 2. Obteniendose el gráfico de módulo, fase, diagráma de polos y ceros, y retardo de grupo.

\begin{equation}
H(s) = \frac{1}{s^2+s\frac{1}{Q}+1}
\tag{1}
\end{equation}

Siendo $Q=\frac{\sqrt{2}}{2}$

```python
# Código para análisis del filtro Butterworth de orden 2
Q_d = np.sqrt(2)/2
w_od = 1
k = -1
num = [k*w_od**2]
den = [1, 1/Q_d, w_od**2]
# Obtención de función transferencia
my_tf = TransferFunction(num,den)

# Gráfico Bode
bodePlot(my_tf, fig_id=1)

# Gráfico de Polos y Ceros
pzmap(my_tf, fig_id=2)

# Group Delay
GroupDelay(my_tf, fig_id=3)
```

#### Simulación Python Butterworth de orden 4
En el siguiente código Python se realiza la simulación de la función transferencia normalizada obtenida en el punto **4)**, para un Filtro Pasa-Bajos Butterworth de orden 4 a partir de dos filtros en cascada, con la topología del ejercicio. Obteniendose el gráfico de módulo, fase, diagráma de polos y ceros, y retardo de grupo.

\begin{equation}
H(s) = \frac{1}{s^2+s\frac{1}{Q_1}+1}\cdot\frac{1}{s^2+s\frac{1}{Q_2}+1}
\tag{1}
\end{equation}

Siendo $Q_1=\frac{1}{2\cdot\cos(\frac{\pi}{8})}$ y $Q_2=\frac{1}{2\cdot\cos(\frac{3}{8}\pi)}$

```python
# Código para análisis del filtro Butterworth de orden 4
Q1 = 1/(2*np.cos(np.pi/8))
wo = 1
k = 1
num1 = [k*wo**2]
den1 = [1, 1/Q1, wo**2]
# Obtención de función transferencia
T1 = TransferFunction(num1,den1)

Q2 = 1/(2*np.cos(3*np.pi/8))
wo = 1
k = 1
num2 = [k*wo**2]
den2 = [1, 1/Q2, wo**2]
# Obtención de función transferencia
T2 = TransferFunction(num2,den2)

my_tf = tfcascade(T1,T2)

# Gráfico Bode
bodePlot(my_tf, fig_id=1)

# Gráfico de Polos y Ceros
pzmap(my_tf, fig_id=2)

# Group Delay
GroupDelay(my_tf, fig_id=3)
```

### Conclusión
Esta configuración de red activa tiene la capacidad de comportarse como filtro **Pasa_Bajos** activo de 2do orden.
Al obtenerse la función transferencia, se observa como el parámetro $Q$ solo depende de $R_2$, mientras que $k$ es inversamente proporcional a $R_1$ y $w_o$ depende de $C$ y $R_3$. Esto la hace una red facilmente configurable, ya que los parámetros característicos dependen de solo uno o dos componentes. Permitiendo de esta manera, lograr filtros de mayor orden rápidamente, como en el punto **4)**.
Por otro lado, se observa la correlación entre las simulaciones de Python y LTspice, teniendo en cuentra que este último genera el bode en función de la frecuencia $f$ y no de la frecuencia angular $w$ como es en el caso de Python.
