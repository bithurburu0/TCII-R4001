Se procede a obtener los polos de la función transferencia y sus respectivos coeficientes $\alpha$ y $\beta$ 

```python
num = [15]
den = [1, 6, 15, 15]
# Se obtienen las raices del denominador de T3(s)
roots = np.roots(den)

alpha1 = np.absolute(np.real(roots[0]))
print("alpha1 = ", alpha1)
beta1 = np.imag(roots[0])
print("beta1 = ", beta1)

alpha2 = np.absolute(np.real(roots[1]))
print("alpha2 = ", alpha2)
beta2 = np.imag(roots[1])
print("beta2 = ", beta2)

alpha3 = np.absolute(np.real(roots[2]))
print("alpha3 = ", alpha3)
beta3 = np.imag(roots[2])
print("beta3 = ", beta3)
```

Gráfico de Retardo de Grupo a partir de la función transferencia de un Bessel Pasa Bajos de orden 3 normalizado

```python
# Gráfico de Retardo de Grupo a partir de la función transferencia de un Bessel Pasa Bajos de orden 3)

w, T = signal.freqs(num, den)
plt.figure()
plt.semilogx(w[1:], -np.diff(np.unwrap(np.angle(T)))/np.diff(w))
plt.grid()
plt.title('Retardo de Grupo para un Bessel LP de orden 3. Método analítico')
plt.show()
```

Obtención de la función transferencia de la configuración Kerwin–Huelsman–Newcomb (KHN), filtro de orden 2 y del Bessel de orden 3 logrado por la concatenación de ambos filtros

```python
# Simulación simbólica

# Transferencia de Kerwin–Huelsman–Newcomb (KHN)
# variables simbólicas
s = sp.symbols('s', complex=True)
Vin, V1, V2, V3, Va = sp.symbols("Vin, V1, V2, V3, Va")
G1, G2, G3, G4, G5, G6, C1, C2 = sp.symbols("G1, G2, G3, G4, G5, G6, C1, C2")

# Sistemas de ecuaciones del modelo ideal, negativamente realimentado
aa = sp.solve([
                V1*G1 + V2*s*C1, 
                V2*G2 + V3*s*C2,
                Va*(G3 + G4) - V1*G4 - V3*G3,
                Va*(G5 + G6) - Vin*G5 - V2*G6
                ], 
                [Vin, V1, V2, V3, Va])
T1 = aa[V3]/aa[Vin]

num, den = sp.fraction(sp.simplify(sp.expand(T1)))
num = sp.Poly(num,s)
den = sp.Poly(den,s)

num1, den1, w_on, Q_n, w_od, Q_d, k = parametrize_sos(num, den)

print_subtitle('Transferencia de Kerwin–Huelsman–Newcomb (KHN)')

display(Math( r' \frac{V_o}{V_i} = ' + sp.latex(sp.Mul(k,num1/den1, evaluate=False)) ))

print_subtitle('Parámetros')

display(Math( r' k = ' + sp.latex(k) ))

display(Math( r' \omega_0 = ' + sp.latex(w_od) ))

display(Math( r' Q = ' + sp.latex(Q_d) ))


# Pasa Bajos de orden 1
# Definición de variables simbólicas
C3, G7 = sp.symbols("C3, G7")

num2 = G7/C3
den2 = s + G7/C3

print_subtitle('Transferencia de pasa bajos orden 1')
display(Math( r' \frac{V_o}{V_i} = ' + sp.latex(sp.Mul(num2/den2, evaluate=False)) ))


# Pasa Bajos Bessel de orden 3
print_subtitle('Transferencia de filtro Pasa Bajos Bessel de orden 3')
display(Math( r' \frac{V_o}{V_i} = ' + sp.latex(sp.simplify(sp.expand(sp.Mul(num1/den1, num2/den2, evaluate=False))))) )
```

Sintetización de la función transferencia para la obtención de la red normalizada y desnormalizada

```python
# Se obtienen los valores de los componentes en función de los parámetros wo y Q

C = 1000e-12 # 1000pF=1nF
C3 = 1000e-12 # 1000pF=1nF
R = 50e3

D0 = 80e-6
omega_f = 1/D0

Wo = np.sqrt(alpha1**2 + beta1**2)
Q = Wo/(2*alpha1)
# Valores normalizados de los componentes luego de la sintetilazación

C3_n = 1
R7_n = Wo**2/15

omega_z1 = R # R=50k
omega_z2 = C3_n/(omega_f*C3)

Cn = omega_f * omega_z1 * C
R1_n = Q/(Wo*Cn)
R2_n = 1/(Wo*Q*Cn)

# Valores desnormalizados de los componentes, para D(0)=80uSeg

R1 = R1_n*omega_z1
R2 = R2_n*omega_z1
R7 = R7_n*omega_z2

print("R1 = ", R1)
print("R2 = ", R2)
print("R7 = ", R7)
print("C3 = ", C3)
```

##### Retardo de Grupo:

```python
def arctan_func(x, a, b, c, d):
    return a * np.arctan(b * x + c) + d

RF2 = np.genfromtxt("OSC_FASE.csv", delimiter=";")

Frec_RF2 = RF2[:,0]
Fase_RF2 = RF2[:,1]

# Calculo del Retardo de Grupo
dphase_dFrec = np.gradient(Fase_RF2*np.pi/180, 2*np.pi*Frec_RF2)
retardo_grupo = -dphase_dFrec

# Ajuste de la curva a los datos
initial_guess = [0.001, 0.001, 0.0, 0.0]  # Asegúrate de usar una estimación inicial razonable
optimizedParameters, covariance = curve_fit(arctan_func, Frec_RF2, retardo_grupo, p0=initial_guess)

# Creación del gráfico de los datos originales
plt.scatter(Frec_RF2, retardo_grupo, label="Datos originales")

# Uso de los parámetros optimizados para trazar la mejor curva de ajuste
plt.plot(Frec_RF2, arctan_func(Frec_RF2, *optimizedParameters), label="Ajuste", color='red')

# Grafico Retardo de Grupo
plt.plot(Frec_RF2, retardo_grupo, label="Retardo de Grupo")
plt.xscale("log")
plt.grid(linestyle='--')
plt.title("Retardo de Grupo en función de la frecuencia medición Osciloscopio")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Retardo de Grupo (s)")
plt.plot(3000, 80e-6-(5*80e-6)/100, marker="o", label="Retardo máx", color="red")
plt.legend()
plt.show()
```
